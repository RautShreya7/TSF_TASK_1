Task 1 :- Prediction Using Supervised ML

Import Necessary Libraries

```python
%matplotlib inline
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = "http://bit.ly/w-data"
data = pd.read_csv(url)
print("Data Imported Succesfully")
type(data)
```

**Printing the Dataset**

```python
data
```

**2D graph to Visualize our Dataset** 

```python
data.plot(x='Hours' , y='Scores', style='o')
plt.title('Hours vs Percentage')
plt.xlabel('Hours Studied')
plt.ylabel('Percentage Score')
plt.show()
```

```python
data.shape
```

```python
X = data.loc[:,['Hours']].values
```

```python
X
```

```python
Y = data.loc[:,'Scores'].values
```

```python
Y
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, random_state=0, test_size=0.2)
```

```python
rg = LinearRegression(fit_intercept=True)
```

```python
rg.fit(X_train, Y_train)
```

```python
rg.predict(X[0].reshape(-1,1))
```

```python
rg.predict(X[0:24])
```

```python
score = rg.score(X_test, Y_test)
```

```python
print(score)
```

```python
p = rg.coef_
p
```

```python
q = rg.intercept_
q
```

```python
#plotting the regression line
line = rg.coef_*X_train+rg.intercept_

plt.scatter(X_train,Y_train)
plt.plot(X_train, line);
plt.show()
```

**Prediction**

```python
pr = rg.predict(X_test)
```

```python
list(zip(Y_test,pr))
```

Model Evaluation

The last step is to evaluate the performance of algorithm.This is an important step and give the comaprison of how well the algorithm perform on a particular dataset. 

```python
from sklearn.metrics import mean_squared_error
```

```python
mean_squared_error(Y_test,pr,squared=False)
```
